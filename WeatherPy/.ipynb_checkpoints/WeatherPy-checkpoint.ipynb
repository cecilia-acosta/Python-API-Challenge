```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import time

from pprint import pprint
from scipy.stats import linregress
import scipy.stats as st

from config import api_key_OpenWeather
from citipy import citipy
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"

query = url + "appid=" + api_key_OpenWeather + "&q="+city
query
```

```python
main_request = requests.get(query)
main_json = main_request.json()

print(main_request)
pprint(main_json)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
query = url + "appid=" + api_key_OpenWeather + "&q="

for city in cities:
    try:
        print(f"Processing Record | {city}")
        response = requests.get(query+city).json()
    except:
        print("City not found. Skipping...")
```

```python
cities_df=pd.DataFrame({
    "City":cities
})
cities_df
```

```python
cities_df["Cloudiness"] = ""
cities_df["Country"] = ""
cities_df["Date"] = ""
cities_df["Humidity"] = ""
cities_df["Lat"] = ""
cities_df["Lng"] = ""
cities_df["Max Temp"] = ""
cities_df["Wind Speed"] = ""

cities_df
```

```python
cities_df.columns
```

```python
for index, row in cities_df.iterrows():
    url = "http://api.openweathermap.org/data/2.5/weather?"
    base_url = url + "appid=" + api_key_OpenWeather + "&q="
    
    city = row['City']
    
    response = requests.get(base_url + city).json()
    
    try:
        print(f'{city} match')
        cities_df.loc[index, 'Cloudiness'] = response['clouds']['all']
        cities_df.loc[index, 'Country'] = response['sys']['country']
        cities_df.loc[index, 'Date'] = response['dt']
        cities_df.loc[index, 'Humidity'] = response['main']['humidity']
        cities_df.loc[index, 'Lat'] = response['coord']['lat']
        cities_df.loc[index, 'Lng'] = response['coord']['lon']
        cities_df.loc[index, 'Max Temp'] = response['main']['temp_max']
        cities_df.loc[index, 'Wind Speed'] = response['wind']['speed']
    except:
        print(f'{city} not found by OpenWeather')
```

```python
cities_df["Cloudiness"]=pd.to_numeric(cities_df["Cloudiness"], downcast="float")
cities_df["Humidity"]=pd.to_numeric(cities_df["Humidity"], downcast="float")
cities_df["Lat"]=pd.to_numeric(cities_df["Lat"], downcast="float")
cities_df["Lng"]=pd.to_numeric(cities_df["Lng"], downcast="float")
cities_df["Max Temp"]=pd.to_numeric(cities_df["Max Temp"], downcast="float")
cities_df["Wind Speed"]=pd.to_numeric(cities_df["Wind Speed"], downcast="float")
cities_df.dtypes
```

## Max Temp

```python
temp=cities_df2[["Hemisphere","Lat","Max Temp"]]
temp_north=temp.loc[temp["Hemisphere"]=="Northern Hemisphere"]
temp_south=temp.loc[temp["Hemisphere"]=="Southern Hemisphere"]
```

```python
x_axis=temp_north["Lat"]
y_axis=temp_north["Max Temp"]
```

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis, y_axis)
regress_values = x_axis * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_axis,y_axis,color="darkblue")
plt.plot(x_axis,regress_values,"r-")
plt.annotate(line_eq,(10,260),fontsize=15,color="red")
plt.title('Northern Hemisphere - Max Temp vs. Latitude Linear Regression')
plt.xlabel('Latitude')
plt.ylabel('Max Temp (F)')
plt.show()
print(rvalue)
```

```python
x_axis=temp_south["Lat"]
y_axis=temp_south["Max Temp"]
```

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis, y_axis)
regress_values = x_axis * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_axis,y_axis,color="darkblue")
plt.plot(x_axis,regress_values,"r-")
plt.annotate(line_eq,(-50,305),fontsize=15,color="red")
plt.title('Southern Hemisphere - Max Temp vs. Latitude Linear Regression')
plt.xlabel('Latitude')
plt.ylabel('Max Temp (F)')
plt.show()
print(rvalue)
```
